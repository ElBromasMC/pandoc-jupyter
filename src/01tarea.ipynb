```python
from sympy import *

N1, N2 = symbols('N1 N2')
b1, b2 = symbols('b1 b2')
r1, r2 = symbols('r1 r2')
k1, k2 = symbols('k1 k2')

dN1 = r1*N1*(1-N1/k1)-b1*r1/k1*N1*N2
dN2 = r2*N2*(1-N2/k2)-b2*r2/k2*N2*N1

X = Matrix([dN1, dN2])
Y = Matrix([N1, N2])

J =simplify(X.jacobian(Y))

N1_, N2_ = (k2*b1-k1)/(b1*b2-1), (k1*b2-k2)/(b1*b2-1)

traza = simplify(J.trace().subs({N1: N1_, N2: N2_}))
determinante = simplify(J.det().subs({N1: N1_, N2: N2_}))
discriminante = simplify(traza**2 - 4*determinante)

J
```
